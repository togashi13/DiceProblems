```python
import numpy as np
```

```python
def states( d, v ):
    return 64 * ( d - 1 ) + v[ 0 ] * 1 + v[ 1 ] * 2 + v[ 2 ] * 4 + v[ 3 ] * 8 + v[ 4 ] * 16 + v[ 5 ] * 32
```

```python
A = np.zeros( ( 385, 385 ) )
for d in range( 1, 7 ):
    for a0 in range( 2 ):
        for a1 in range( 2 ):
            for a2 in range( 2 ):
                for a3 in range( 2 ):
                    for a4 in range( 2 ):
                        for a5 in range( 2 ):
                            v = np.asarray( [ a0, a1, a2, a3, a4, a5 ] )
                            state_ = states( d, v )
                            for i in range( 1, 7 ):
                                m = np.asarray( [ a0, a1, a2, a3, a4, a5 ] )
                                m[ abs( d - i ) ] = 1
                                if all( m == np.ones( 6 ) ):
                                    A[ state_, -1 ] = 1 
                                else:
                                    state_x = states( i, m )
                                    A[ state_, state_x ] += 1 / 6
```

```python
a = np.linalg.inv( np.identity( 384 ) - A[ : -1, : -1 ] )
a
```

```python
g = a @ (np.ones( ( 384, 1 ) ) )
g
```
