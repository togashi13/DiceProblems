```python
import numpy as op
```

```python
def rolls_to_state( v ):
    return 1 + v[ 0 ] + v[ 1 ] * 3 + v[ 2 ] * 9 + v[ 3 ] * 27 + v[ 4 ] * 81 + v[ 5 ] * 243 
```

```python
rolls_to_state( [ 2, 0, 0, 0, 0, 0 ] )
```

```python
A = op.zeros( ( 730, 730 ) )
for a1 in range( 3 ):
    for a2 in range( 3 ):
        for a3 in range( 3 ):
            for a4 in range( 3 ):
                for a5 in range( 3 ):
                    for a6 in range( 3 ):
                        v = op.array( [ a1, a2, a3, a4, a5, a6 ] )
                        s = rolls_to_state( v )
                        c = 0
                        for _ in v:
                            if _ == 2:
                                c += 1
                        A[ s, 0 ] = c / 6
                        if s == 3:
                            print( c )
                        for i in range( len( v ) ):
                            w = op.array( [ a1, a2, a3, a4, a5, a6 ] )
                            if w[ i ] < 2:
                                w[ i ] += 1
                                ss = rolls_to_state( w )
                                A[ s, ss ] = 1 / 6
```
